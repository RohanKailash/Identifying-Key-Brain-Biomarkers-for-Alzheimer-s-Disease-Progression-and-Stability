## Dataset Retrieval

The dataset we take consist of 39 Brain Biomarkers for both left and right Brain and their respective baseline and current Diagnosis. The initial diagnosis is of 5 catogories:

#### Cognitively Normal Subjects(CN):
MMSE scores between 24-30 (inclusive), a CDR of 0, non-depressed, non-MCI, and non-demented, education adjusted scores on delayed recall of one paragraph from Wechsler Memory Scale Logical Memory II ( ≥9 for 16 or more years of education, ≥5 for 8-15 years of education, ≥3 for 0-7 years of education).

#### Significant Memory Concern Subjects(SMC): 
MMSE scores between 24-30 (inclusive), a significant subjective memory concern reported by subject, informant, or clinician, CCI score ≥16 (based on first 12 questions), a CDR of 0, non-depressed, non-MCI, and non-demented, education adjusted scores on delayed recall of one paragraph from Wechsler Memory Scale Logical Memory II (≥9 for 16 or more years of education, ≥5 for 8-15 years of education, ≥3 for 0-7 years of education).

#### EMCI Subjects: 
MMSE scores between 24-30 (inclusive), a subjective memory concern reported by subject, informant, or clinician, objective memory loss measured by education adjusted scores on delayed recall of one paragraph from Wechsler Memory Scale Logical Memory II (≥16 years: 9-11; 8-15 years: 5-9; 0-7 years: 3-6), a CDR of 0.5, absence of significant levels of impairment in other cognitive domains, essentially preserved activities of daily living, and an absence of dementia.

#### LMCI Subjects: 
MMSE scores between 24-30 (inclusive), a subjective memory concern reported by subject, informant, or clinician, objective memory
loss measured by education adjusted scores on delayed recall of one paragraph from Wechsler Memory Scale Logical Memory II (≥16 years: ≤8; 8-15 years: ≤4; 0-7 years: ≤2), a CDR of 0.5, absence of significant levels of impairment in other cognitive domains, essentially preserved activities of daily living, and an absence of dementia.

#### Mild AD Subjects(MCI): 
MMSE scores between 20-26 (inclusive), a CDR of 0.5 or 1.0, and meets NINCDS/ADRDA criteria for probable AD

#### Alzheimer's Disease(AD):
Refers to patients who have been diagnosed with Alzheimer's disease, characterized by progressive cognitive decline and memory loss.

### Cleaning the Dataset

#### Drop unwanted columns

```python
# List of columns to drop
columns_to_drop = [
    "ORIGPROT", "COLPROT", "MagStrength", "Sequence", "ScanDate", 
    "StudyID", "SeriesID", "ImageUID", "Description", "FilesBase","SegmentationID"
]

# Drop the columns
brain_bio_df = brain_bio_df.drop(columns=columns_to_drop, errors="ignore")

# Display the modified dataframe
print(brain_bio_df.head())
```

#### Merge Age at baseline and Age during scan to capture after how long progress or No progress happens

```python
# Create the new column "Age_diff"
brain_bio_df["Age.diff"] = brain_bio_df["AGE.scan"] - brain_bio_df["AGE.bl"]

# Drop the original columns
brain_bio_df = brain_bio_df.drop(columns=["AGE.bl", "AGE.scan"], errors="ignore")

# Display the first few rows
print(brain_bio_df.head())
```

#### Find difference all biomarkers from Baseline

```python
# Define columns that should not be modified
columns_to_keep = ["RID", "VISCODE", "MMSE", "DX.bl", "DX.scan"]

# Function to process the DataFrame
def adjust_rows(df):
    df_bl = df[df["VISCODE"] == "bl"]  # Filter rows where VISCODE is "bl"
    df_non_bl = df[df["Age.diff"] != 0]  # Rows where Age_diff is not zero
    
    adjusted_rows = []
    
    for _, row in df_non_bl.iterrows():
        # Find the corresponding baseline (R1) row
        R1 = df_bl[df_bl["RID"] == row["RID"]]
        
        if not R1.empty:
            R1 = R1.iloc[0]  # Get the first matching baseline row
            
            # Copy the row to modify
            new_row = row.copy()
            
            # Subtract values from R1 (excluding columns_to_keep)
            for col in df.columns:
                if col not in columns_to_keep:
                    new_row[col] = row[col] - R1[col]
            
            adjusted_rows.append(new_row)
    
    # Create a new DataFrame with adjusted rows
    final_df = pd.DataFrame(adjusted_rows)
    
    return final_df
```

```python
# Apply the function
brain_bio_df = adjust_rows(brain_bio_df)

# Display result
print(brain_bio_df.head())
```

#### Labelling Progression
CN -> SMC -> EMCI -> LMCI ->MCI ->AD [According to ADNI manual for New Patients]

Classes
1. No progress: 316+889 = 1205
2. Positive Progress(Towards Dementia): 20+100+23+354+621+1009 = 2127
3. Negative Progress(Towards Normal): 3+22+43 = 68

```python
def classify_progress(row):
    dx_bl, dx_scan = row['DX.bl'], row['DX.scan']
    
    if (dx_bl, dx_scan) in [('AD', 'AD'), ('CN', 'CN')]:
        return "No progress"
    elif (dx_bl, dx_scan) in [('CN', 'AD'), ('CN', 'MCI'), ('EMCI', 'AD'), ('EMCI', 'MCI'), ('LMCI', 'AD'), ('LMCI', 'MCI')]:
        return "Positive progress"
    elif (dx_bl, dx_scan) in [('AD', 'MCI'), ('EMCI', 'CN'), ('LMCI', 'CN')]:
        return "Negative progress"
    else:
        return "Unknown"

# Assuming brain_bio_df is already loaded
brain_bio_df['Progress'] = brain_bio_df.apply(classify_progress, axis=1)
```

```python
#DFucntion to train model with given df
def train_models(X,y):
# Initialize models
    models = {
        'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
        'SVM': SVC(random_state=42),
        'Random Forest': RandomForestClassifier(random_state=42),
        'Decision Tree': DecisionTreeClassifier(random_state=42)
    }

    # Dictionary to store results
    results = {}

    # Implement and evaluate each model
    for name, model in models.items():
        print(f"\n{'-'*50}")
        print(f"Implementing {name}")
        
        # Perform 10-fold cross-validation
        cv_scores = cross_val_score(model, X, y, cv=10, scoring='accuracy')
        
        # Print detailed results
        print(f"Cross-validation scores: {cv_scores}")
        print(f"Mean CV accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
        
        # Store results
        results[name] = cv_scores.mean()
        
        # Fit model on full dataset for feature importance (if applicable)
        model.fit(X, y)
        
        # Print feature importance for Random Forest and Decision Tree
        if name in ['Random Forest', 'Decision Tree']:
            importances = pd.DataFrame({
                'Feature': X.columns,
                'Importance': model.feature_importances_
            }).sort_values('Importance', ascending=False)
            print("\nFeature Importance:")
            print(importances)
            importances.to_csv(f"{name} Importance.csv")

    # Print summary of results
    print(f"\n{'-'*50}")
    print("Summary of Model Performances:")
    for name, score in results.items():
        print(f"{name}: {score:.4f}")
```

## Using Class weights to avoid class imbalance

Class weights offer a simple yet effective technique to deal with class imbalance. The idea is to assign higher weights to the samples of the minority class and lower weights to the majority class during the training process. By doing this, the model pays more attention to the minority class and learns to make better predictions for it.

```python
target_column='Progress'
X = df.drop(target_column, axis=1)
y = df[target_column]
```

```python
# Calculating class weights using Inverse class Frequency method
import numpy as np

def calculate_class_weights(y):
    unique_classes, class_counts = np.unique(y, return_counts=True)
    total_samples = len(y)
    class_weights = {}

    for class_label, class_count in zip(unique_classes, class_counts):
        class_weight = total_samples / (2.0 * class_count)
        class_weights[class_label] = class_weight

    return class_weights
```

```python
#DFucntion to train model with given df
def train_models_weight_balanced(X,y,class_weights):
# Initialize models
    models = {
        'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42,class_weight=class_weights),
        'SVM': SVC(random_state=42,class_weight=class_weights),
        'Random Forest': RandomForestClassifier(random_state=42,class_weight=class_weights),
        'Decision Tree': DecisionTreeClassifier(random_state=42,class_weight=class_weights)
    }

    # Dictionary to store results
    results = {}

    # Implement and evaluate each model
    for name, model in models.items():
        print(f"\n{'-'*50}")
        print(f"Implementing {name}")
        
        # Perform 10-fold cross-validation
        cv_scores = cross_val_score(model, X, y, cv=10, scoring='accuracy')
        
        # Print detailed results
        print(f"Cross-validation scores: {cv_scores}")
        print(f"Mean CV accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
        
        # Store results
        results[name] = cv_scores.mean()
        
        # Fit model on full dataset for feature importance (if applicable)
        model.fit(X, y)
        
        # Print feature importance for Random Forest and Decision Tree
        if name in ['Random Forest', 'Decision Tree']:
            importances = pd.DataFrame({
                'Feature': X.columns,
                'Importance': model.feature_importances_
            }).sort_values('Importance', ascending=False)
            print("\nFeature Importance:")
            print(importances)
            importances.to_csv(f"{name} Importance.csv")

    # Print summary of results
    print(f"\n{'-'*50}")
    print("Summary of Model Performances:")
    for name, score in results.items():
        print(f"{name}: {score:.4f}")
```

```python
class_weights=calculate_class_weights(y)
print(class_weights)
train_models_weight_balanced(X,y,class_weights)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier

def train_model(X_train,y_train,X_test,y_test):
    
    class_labels = np.unique(y_test)

    model =RandomForestClassifier(random_state=42)
    model.fit(X_train,y_train)

    y_pred = model.predict(X_test)
    # Classification report (precision, recall, f1-score per class)
    print("Classification Report:")
    print(classification_report(y_test, y_pred, target_names=class_labels))

    # Confusion matrix
    conf_matrix = confusion_matrix(y_test, y_pred)
    print("\nConfusion Matrix:")
    print(conf_matrix)

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title("Confusion Matrix")
    plt.show()

    # Compute per-class accuracy
    class_accuracy = conf_matrix.diagonal() / conf_matrix.sum(axis=1)
    
    # Print accuracy per class
    print("\nPer-Class Accuracy:")
    for i, label in enumerate(class_labels):
        print(f"{label}: {class_accuracy[i]:.4f}")

    return class_accuracy
```
